```python
import torch, pickle, metrics
import numpy as np
```

```python
infersent_encoding_path = '../data/s2_corpus_infersent_fasttext.pkl'

infersent_data = {}
with open(infersent_encoding_path+'.1', 'rb') as f:
    data = pickle.load(f)
    for k in data:
        infersent_data[k] = data[k]
with open(infersent_encoding_path+'.2', 'rb') as f:
    data = pickle.load(f)
    for k in data:
        infersent_data[k] = data[k]
```

```python
triplet_path = '../data/../data/s2_triplets.pkl'
with open(triplet_path ,'rb') as f:
    triplet_data = pickle.load(f)
```

```python
split = 'test'
split_data = triplet_data[split]
ancs = np.array(
    [infersent_data[id] for id, _, _ in split_data]
)
poss = np.array(
    [infersent_data[id] for _, id, _ in split_data]
)
negs = np.array(
    [infersent_data[id] for _, _, id in split_data]
)
```

```python
def num_is_correct_eucl(ancs_encoding, poss_encoding, negs_encoding):
    distance_positive = ((ancs_encoding - poss_encoding)**2).sum(1)
    distance_negative = ((ancs_encoding - negs_encoding)**2).sum(1)
    # (B, )
    num_is_correct_eucl = (distance_positive < distance_negative).sum()
    return num_is_correct_eucl
```

```python
num_is_correct_eucl(ancs, poss, negs) / len(split_data)
```
